Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Five common plots using matplotlib.pyplot:

Line Plot – For visualizing trends over time or ordered data.

Bar Chart – To compare quantities across different categories.

Histogram – For showing the distribution of a dataset.

Scatter Plot – To observe relationships or correlations between two variables.

Pie Chart – For showing parts of a whole as percentages.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(3)

# Generate data
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot
plt.scatter(x, y, color='blue', marker='o')  # scatter plot of x vs y
plt.title('Scatter Plot of Randomly Generated Data')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for all lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots (2 rows, 2 columns)
plt.figure(figsize=(10, 8))  # Optional: size of the whole figure

plt.subplot(2, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, color='blue')
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, color='green')
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, color='purple')
plt.title('Line 4')

# Automatically adjusts layout so titles and plots don't overlap
plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color='skyblue')
plt.title("Company Profits (Vertical)")
plt.xlabel("Company")
plt.ylabel("Profit ($ millions)")

# Horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color='lightgreen')
plt.title("Company Profits (Horizontal)")
plt.xlabel("Profit ($ millions)")
plt.ylabel("Company")

plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(0)  # for reproducibility
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Plot box plot
data = [box1, box2]

plt.boxplot(data, labels=["Box 1", "Box 2"], patch_artist=True,
            boxprops=dict(facecolor="lightblue"), medianprops=dict(color="red"))

plt.title("Box Plot Comparison")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```
