Q1.What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases

What is MongoDB?

MongoDB is an open-source, non-relational (NoSQL) database management system that stores data in flexible, JSON-like documents called BSON.

Non-Relational Databases Explained (in short)

Non-relational databases, often referred to as NoSQL databases, do not use the fixed table structure of relational databases.

When to Prefer MongoDB Over SQL Databases

MongoDB is preferred over traditional SQL (relational) databases in scenarios where:

Schema Flexibility is Needed: The data structure may evolve over time, or records do not all share the same fields (e.g., user profiles with optional attributes).

High Scalability and Big Data: Applications must handle large volumes of data and require horizontal scaling across distributed systems, such as in IoT, analytics, or real-time big data processing.

Q2. State and Explain the features of MongoDB.

Features of MongoDB

Document-Oriented Storage

MongoDB stores data in documents using a flexible, JSON-like format called BSON.

Schema-Less Architecture

MongoDB collections do not require a predefined schema.

Rich Query Language

MongoDB supports a powerful and flexible query language that allows for field queries, range queries, regular expression searches, and aggregation operations.

Q4. Using the database and the collection created in question number 3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted record.

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['school']
students_collection = db['students']

# Insert one record
student_one = {"name": "Alice", "age": 20, "grade": "A"}
students_collection.insert_one(student_one)

# Insert many records
students_many = [
    {"name": "Bob", "age": 21, "grade": "B"},
    {"name": "Charlie", "age": 22, "grade": "A"},
    {"name": "Diana", "age": 20, "grade": "C"}
]
students_collection.insert_many(students_many)

# Find one
print("One student record:")
print(students_collection.find_one())

# Find all
print("All student records:")
for student in students_collection.find():
    print(student)
```

Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

```python
from pymongo import MongoClient

# Connect to MongoDB (default host and port)
client = MongoClient('mongodb://localhost:27017/')

# Select the database and collection
db = client['your_database_name']
users_collection = db['users']

# Query: Find users with age > 25, return only name and age (exclude _id)
results = users_collection.find(
    {'age': {'$gt': 25}},
    {'name': 1, 'age': 1, '_id': 0}
)

# Print the results
for user in results:
    print(user)
```

Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

```python
from pymongo import MongoClient

# Connect to MongoDB (default host and port)
client = MongoClient('localhost', 27017)

# Access the database and collection
db = client['testdb']
collection = db['mycol']

# Insert sample documents (if not already present)
sample_docs = [
    { "_id": 1, "title": "MongoDB Overview" },
    { "_id": 2, "title": "NoSQL Overview" },
    { "_id": 3, "title": "Tutorials Point Overview" }
]
collection.delete_many({})  # Clear collection for demo
collection.insert_many(sample_docs)

# Sort by 'title' in descending order
cursor = collection.find({}, { "_id": 0, "title": 1 }).sort("title", -1)

print("Sorted by title (descending):")
for doc in cursor:
    print(doc)

# Sort by 'title' in ascending order
cursor = collection.find({}, { "_id": 0, "title": 1 }).sort("title", 1)

print("\nSorted by title (ascending):")
for doc in cursor:
    print(doc)
```
