Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px
titanic=sns.load_dataset('titanic')
```

```python
fig = px.scatter(titanic, x="age", y="fare", title="Titanic: Age vs Fare")
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px

# Load the tips dataset
df = px.data.tips()

# Create a simple histogram
fig = px.histogram(df, x='sex',y='total_bill',pattern_shape='smoker',color='day')
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

What is a Distplot?

A distplot (distribution plot) is a visualization that combines several statistical representations of numerical data, typically including:

A histogram (showing the frequency distribution of data)

A curve (such as a kernel density estimation or a normal distribution curve)

A rug plot (showing individual data points along the axis)

Distplots are useful for visualizing the distribution and density of a dataset, making it easier to understand its underlying structure and spread.

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Generate random data
x = np.random.randn(500)
df = pd.DataFrame({'x': x})

# Create histogram with density and rug plot
fig = px.histogram(df, x="x", histnorm="density", marginal="rug", opacity=0.7)

# Add a density curve using KDE
fig.add_trace(
    px.density_contour(df, x="x").data[0]
)

fig.update_layout(title="Approximate Distplot using Plotly Express")
fig.show()
```
